```python
import pandas as pd
import ta
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('data/AAPL.csv', sep=',')

# Clean nan values

# df = ta.utils.dropna(df)
```

```python
# df = ta.utils.dropna(df)

print(df.columns)
```

```python
df = ta.add_all_ta_features(df, "Open", "High", "Low", "Close", "Volume", fillna=True)
```

```python
print(df.columns)
print(len(df.columns))
```

```python
plt.plot(df[500:1000].Close)
plt.plot(df[500:1000].volatility_bbh, label='High DC')
plt.plot(df[500:1000].volatility_bbh, label='Low DC')
plt.title('volatility_bbh')
plt.legend()
plt.show()
```

```python
plt.plot(df[500:1000].Close)
plt.plot(df[500:1000].trend_macd_signal, label='High DC')
plt.plot(df[500:1000].trend_macd_signal, label='Low DC')
plt.title('volatility_bbh')
plt.legend()
plt.show()
```
