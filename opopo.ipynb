```python
import gym
import json
import datetime as dt
import numpy as np

from stable_baselines.common.policies import MlpPolicy
from stable_baselines.common.vec_env import DummyVecEnv
from stable_baselines import PPO2

from env.StockTradingEnv import StockTradingEnv

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/AAPL.csv')
df = df.sort_values('Date')
```

```python
df.head()
```

```python
df_train = df.iloc[:4000]
env_train = DummyVecEnv([lambda: StockTradingEnv(df_train)])
```

```python
model = PPO2(MlpPolicy, env_train, verbose=0,tensorboard_log="./PPO2_tensorboard/")
model.learn(total_timesteps=100000)
```

```python
df_test = df.iloc[4000:]
```

```python
env_test = DummyVecEnv([lambda: StockTradingEnv(df_test)])
```

```python
obs = env_test.reset()
balance_plot = list()
share_plot = list()
profit_plot = list()

obs_plot = list()
n = 60
for i in range(n):
    action, _states = model.predict(obs)
    obs, rewards, done, info = env_test.step(action)
    obs_plot.append(obs)

    balance, profit, share = env_test.render()
    balance_plot.append(balance)
    share_plot.append(share)
    profit_plot.append(profit)
    if done:
        break
```
